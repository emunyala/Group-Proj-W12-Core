```python
import pandas as pd
import re
import numpy as np
```

```python
# clean the text
def clean_text(text):
    # remove the @mentions
    text = re.sub(r'@[A-Za-z0-9]+', '', text)
    # remove the '#' symbol
    text = re.sub(r'#', '', text)
    # remove the ':' symbol
    text = re.sub(r':', '', text)
    # remove the RT
    text = re.sub(r'RT[\s]+','', text)

    #remove backline
    text = re.sub(r'RT[\n]+','', text)

    # remove hyper links
    text = re.sub(r'https?:\/\/\S+', '', text)
    return text

# # apply the function to the DataFrame containing the tweets
# new_tweets['tweets'] = new_tweets['tweets'].apply(clean_text)
# # display the data
# new_tweets.head()
```
